```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('lattice_constants.txt', sep = ' ')
df.head()
```

```python
fig, ax = plt.subplots(figsize = (20,20))
ax.scatter(df['ID'], df['Lattice_Constant'])
ax.errorbar(df['ID'], df['Lattice_Constant'], xerr = None, yerr = df['Error'], capsize = 5, linestyle = ' ')
plt.savefig('LineCut_series.png')
```
