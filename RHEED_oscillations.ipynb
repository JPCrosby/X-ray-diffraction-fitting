```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
```

```python
df = pd.read_csv('Run1.txt', sep = '\t')
df.columns = ['Time', 'Intensity']
df.head()

data = df[df['Time']==0]
print(data)

intensity1 = df[:39583]
intensity2 = df[39584:79167]
intensity3 = df[79168:]
```

```python
fig, ax = plt.subplots(figsize = (20, 10))
intensity1.plot(kind = 'line', x = 'Time', ax=ax)
intensity2.plot(kind = 'line', x = 'Time', ax=ax)
intensity3.plot(kind = 'line', x = 'Time', ax=ax)
ax.set_xlim(1700, df.iloc[-1]['Time'])
ax.set_ylim(5000, 10000)
ax.minorticks_on()
ax.grid(which = 'major', linestyle='-')
ax.grid(which='minor', linestyle=':')
plt.savefig('GM1-606_RHEED_oscillations_endgrowth.pdf')
```
